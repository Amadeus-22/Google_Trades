```python

# Create locators for ticks on the time axis

import matplotlib.pyplot as plt
import matplotlib.dates as mdates


# Create locators for ticks on the time axis
plt.gca().xaxis.set_major_locator(mdates.YearLocator())  # Major ticks every year
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))  # Format as Year
plt.gca().xaxis.set_minor_locator(mdates.MonthLocator())  # Minor ticks every month
```

```python
import matplotlib.pyplot as plt

# Create the figure and first axis
fig, ax1 = plt.subplots(figsize=(14, 8), dpi=100)

# Plot Tesla Web Search on ax1
ax1.plot(df_tesla['MONTH'], df_tesla['TSLA_WEB_SEARCH'], linewidth=3)
ax1.set_xlabel('Date', fontsize=14)
ax1.set_ylabel('Search Trend', fontsize=14)
ax1.tick_params(axis='y')
ax1.tick_params(axis='x', rotation=45)

# Create second axis that shares the same x-axis
ax2 = ax1.twinx()

# Plot Tesla Stock Price on ax2
ax2.plot(df_tesla['MONTH'], df_tesla['TSLA_USD_CLOSE'], linewidth=3)
ax2.set_ylabel('TSLA Stock Price', fontsize=14)
ax2.tick_params(axis='y')

# Add title and grid
plt.title('Tesla Web Search vs Stock Price', fontsize=16)
ax1.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create the figure and first axis
fig, ax1 = plt.subplots(figsize=(14, 8), dpi=120)

# Plot Bitcoin search trend with marker='o' for circles
ax1.plot(df_btc_search['MONTH'], df_btc_search['BTC_NEWS_SEARCH'], 
         color='royalblue', linewidth=2, marker='o', label='Search Trend')
ax1.set_xlabel('Date', fontsize=14)
ax1.set_ylabel('Search Trend', color='royalblue', fontsize=14)
ax1.tick_params(axis='y', labelcolor='royalblue')
ax1.tick_params(axis='x', rotation=45)
ax1.set_ylim([0, df_btc_search['BTC_NEWS_SEARCH'].max() + 10])  

# Create second y-axis for BTC price
ax2 = ax1.twinx()
ax2.plot(df_btc_monthly.index, df_btc_monthly['CLOSE'], 
         color='crimson', linestyle='--', linewidth=2, label='BTC Price')
ax2.set_ylabel('BTC Price', color='crimson', fontsize=14)
ax2.tick_params(axis='y', labelcolor='crimson')
ax2.set_ylim([0, df_btc_monthly['CLOSE'].max() + 1000])  

# Title and grid
plt.title('Bitcoin News Search vs Resampled Price', fontsize=16)
ax1.grid(True, linestyle='--', alpha=0.5)
```

```python
import matplotlib.pyplot as plt

# Create the figure and first axis
fig, ax1 = plt.subplots(figsize=(14, 8), dpi=120)

# Plot unemployment benefits search trend
ax1.plot(df_unemployment['MONTH'], df_unemployment['UE_BENEFITS_WEB_SEARCH'], 
         color='royalblue', linewidth=2, marker='o', label='Search Trend')
ax1.set_xlabel('Date', fontsize=14)
ax1.set_ylabel('Search Trend', color='royalblue', fontsize=14)
ax1.tick_params(axis='y', labelcolor='royalblue')
ax1.tick_params(axis='x', rotation=45)
ax1.set_ylim([0, df_unemployment['UE_BENEFITS_WEB_SEARCH'].max() + 10])  

# Create second axis for unemployment rate
ax2 = ax1.twinx()
ax2.plot(df_unemployment['MONTH'], df_unemployment['UNRATE'], 
         color='crimson', linestyle='--', linewidth=2, label='Unemployment Rate')
ax2.set_ylabel('FRED U/E Rate', color='crimson', fontsize=14)
ax2.tick_params(axis='y', labelcolor='crimson')
ax2.set_ylim([0, df_unemployment['UNRATE'].max() + 2])  

# Add title and dashed gray grid
plt.title('Monthly Search of "Unemployment Benefits" in the U.S. vs the U/E Rate', fontsize=16)
ax1.grid(True, linestyle='--', color='gray', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
df_unemployment['6MA_WEB_SEARCH'] = df_unemployment['UE_BENEFITS_WEB_SEARCH'].rolling(window=6).mean()

import matplotlib.pyplot as plt

# Create figure and primary axis
fig, ax1 = plt.subplots(figsize=(14, 8), dpi=120)

# Plot 6-month rolling average
ax1.plot(df_unemployment['MONTH'], df_unemployment['6MA_WEB_SEARCH'], 
         color='royalblue', linewidth=2, label='6-Month Rolling Avg (Search)')
ax1.set_xlabel('Date', fontsize=14)
ax1.set_ylabel('6-Month Search Trend', color='royalblue', fontsize=14)
ax1.tick_params(axis='y', labelcolor='royalblue')
ax1.tick_params(axis='x', rotation=45)

# Create secondary axis
ax2 = ax1.twinx()
ax2.plot(df_unemployment['MONTH'], df_unemployment['UNRATE'], 
         color='crimson', linestyle='--', linewidth=2, label='Unemployment Rate')
ax2.set_ylabel('FRED U/E Rate', color='crimson', fontsize=14)
ax2.tick_params(axis='y', labelcolor='crimson')

# Title and grid
plt.title('6-Month Rolling Average of Web Searches vs Unemployment Rate', fontsize=16)
ax1.grid(True, linestyle='--', color='gray', alpha=0.6)

plt.tight_layout()
plt.show()
```
